# Data Visualization

```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[4,5,6]
plt.plot(x,y)
plt.title('New Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')

x1=[1,2,3]
y1=[4,5,6]

x2=[6,7,8]
y2=[9,10,7]

plt.plot(x1,y1,'g',label='first',linewidth=5)
plt.plot(x2,y2,'c',label='second',linewidth=5)

plt.title('New Graph')

plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.legend()

plt.grid(True,color='k')
plt.show()
```

# BAR GRAPH

```python
import matplotlib.pyplot as plt
x1=[1,3,5,7]
y1=[5,2,6,6]

x2=[2,4,6,8]
y2=[6,3,5,2]

plt.bar(x1,y1,label='first bar',color='red')
plt.bar(x2,y2,label='second bar',color='black')

plt.legend()

plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.title('Bar Graph')

plt.show()
```

# HISTOGRAM

```python
import matplotlib.pyplot as plt

Sale_values=[12,10,5,7,0,14,15,24,10,11,15,30,18,11,10,18,30]

bins=[0,5,10,15,20,25,30]
plt.hist(Sale_values,bins,histtype='bar',rwidth=.6)

plt.xlabel('X axis',color='g')
plt.ylabel('Y axis',color='k')

plt.title('Histogram Graph')

plt.show()
```

# SCATTER GRAPH(imp)

```python
import matplotlib.pyplot as plt

x=[1,3,4,8,5,6,7]
y=[3,5,6,7,8,9,3]

plt.scatter(x,y)

plt.xlabel('X axis',color='g')
plt.ylabel('Y axis',color='r')

plt.title('Scatter Graph')

plt.show()
```

# STACKPLOT

```python
days=[1,2,3,4,5]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]

plt.plot([],[],color='m',label='sleeping',linewidth=5)
plt.plot([],[],color='m',label='eating',linewidth=5)
plt.plot([],[],color='m',label='working',linewidth=5)
plt.plot([],[],color='m',label='playing',linewidth=5)

plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','k'])
plt.title('Interesting Graph check it out')

plt.legend()
plt.show()
```

# PIE CHART

```python
import matplotlib.pyplot as plt
labels= 'python','c++','Ruby','Java'
sizes=[215,130,245,210]
colors=['b','c','r','g']
explode=(.1,0,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,
       autopct='%1.1f%%',shadow=True,startangle=140)
plt.axis('equal')
plt.show()
```
