# merging of the data

```python
import pandas as pd
df1=pd.DataFrame({"emp_name":[10,14,16,17],"project":[2,4,5,7],"Revenue":[10,20,45,80]},index=[2001,2002,2003,2004])
df2=pd.DataFrame({"emp_name":[10,14,16,17],"project":[2,4,5,7],"Revenue":[10,20,45,80]},index=[2005,2006,2007,2008])
merge=pd.merge(df1,df2)
merge
```

```python
import pandas as pd
df1=pd.DataFrame({"emp_name":[10,14,16,17],"project":[2,4,5,7],"Revenue":[10,20,45,80]},index=[2001,2002,2003,2004])
df2=pd.DataFrame({"emp_name":[10,14,16,17],"project":[2,4,5,8],"Revenue":[10,20,45,80]},index=[2005,2006,2007,2008])
merge=pd.merge(df1,df2)
merge
```

```python
import pandas as pd
df1=pd.DataFrame({"emp":[10,14,16,17],"project":[2,4,5,7],"Revenue":[10,20,45,80]},index=[2001,2002,2003,2004])
df2=pd.DataFrame({"emp":[10,14,16,17],"project":[2,4,5,7],"Revenue":[10,20,45,80]},index=[2005,2006,2007,2008])
merge=pd.merge(df1,df2,on="emp")
merge
```

# Joining of the data

```python
import pandas as pd
df1=pd.DataFrame({"emp":[10,14,16,17],"project":[2,4,5,7]},index=[2001,2002,2003,2004])
df2=pd.DataFrame({"leaves":[40,28,44,22],"Revenue":[10,20,45,80]},index=[2001,2002,2004,2005])
joined=df1.join(df2)
joined
```

```python
import pandas as pd
df1=pd.DataFrame({"emp":[10,14,16,17],"project":[2,4,5,7]},index=[2001,2002,2003,2004])
df2=pd.DataFrame({"leaves":[40,28,44,22],"Revenue":[10,20,45,80]},index=[2001,2002,2004,2005])
joined=df2.join(df1)
joined
```

##  concatenate

```python
df1=pd.DataFrame({"emp":[10,14,16,17],"project":[2,4,5,7]},index=[2001,2002,2003,2004])
df2=pd.DataFrame({"emp":[10,14,16,17],"project":[2,4,5,7]},index=[2005,2006,2007,2008])
concatenate=pd.concat([df1,df2])
concatenate
```
